## Question 1

## Question 2
1. Load in `positive_words.txt` from `data/`.
2. Find a list of words that contains only 1 vowel, 2 vowels, ..., 4 vowels.
    - What is the average number of vowels in the list of positive words?
3. Find a list of positive adverbs ending with `ly`. How many are there?
4. Find a list of positive words beginning with `re`. How many are there?
5. Find a list of positive verbs ending in `ing`. How many are there?


```python
positive_words = open("../data/positive_words.txt", encoding='utf-8').read()
positive_words = positive_words.split('\n')
positive_words = [word for word in positive_words if "-" not in word]
positive_words
```

```python
vowel1 = [re.findall(r"\b[^aeiou]*[aeiou]{1}[^aeiou]*\b", word) for word in positive_words if len(re.findall(r"\b[^aeiou]*[aeiou]{1}[^aeiou]*\b", word)) > 0]
vowel1
```

```python
vowel2 = [re.findall(r"\b[^aeiou]*[aeiou]{2}[^aeiou]*\b", word) for word in positive_words if len(re.findall(r"\b[^aeiou]*[aeiou]{2}[^aeiou]*\b", word)) > 0]
vowel2
```

```python
vowel3 = [re.findall(r"\b[^aeiou]*[aeiou]{3}[^aeiou]*\b", word) for word in positive_words if len(re.findall(r"\b[^aeiou]*[aeiou]{3}[^aeiou]*\b", word)) > 0]
vowel3
```

```python
vowel4 = [re.findall(r"\b[^aeiou]*[aeiou]{4}[^aeiou]*\b", word) for word in positive_words if len(re.findall(r"\b[^aeiou]*[aeiou]{4}[^aeiou]*\b", word)) > 0]
vowel4
```

```python
average_vowels = (1*len(vowel1)+2*len(vowel2)+3*len(vowel3)+4*len(vowel4)) / (len(vowel1)+len(vowel2)+len(vowel3)+len(vowel4))
average_vowels
```

```python
ly_words = [re.findall(r'\w*ly$', word) for word in positive_words if len(re.findall(r'\w*ly$', word)) > 0]
ly_words
```

```python
len(ly_words)
```

```python
re_words = [re.findall(r're\w*$', word) for word in positive_words if len(re.findall(r're\w*$', word)) > 0]
re_words
```

```python
len(re_words)
```

```python
ing_words = [re.findall(r'\w*ing$', word) for word in positive_words if len(re.findall(r'\w*ing$', word)) > 0]
ing_words
```

```python
len(ing_words)
```
